# Crear visualizaciones

Luego creamos un dashboard en el que pondremos nuestros graficos, metricas, mapas, todo lo que nos permite kibana, en este caso usaremos algunas metricas y un grafico de linea simple como un pequeño ejemplo.

+ Creamos Dashboard
+ Creamos una visualización

1. KPI 1: Promedio Nacional
- Cambiamos el tipo a Metric

3. KPI 3: Precio Mínimo Registrado
- Create visualization
- Tipo → Metric
