# ¿ Qué es el Stack ELK ?

Stack compuesta por tres pilares:

**ElasticSearch**, **Logstash** y **Kibana**

- Open Source
- Desarrolladas y mantenidas por la empresa Elastic

### Elastic Search

Proporciona  la capacidad de almacenamiento y de búsqueda de datos en el sitema

### Logstash

Recopila datos de múltiples fuentes los transforma, limpia y estructura en tiempo real
