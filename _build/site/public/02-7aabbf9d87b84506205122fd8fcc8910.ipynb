# Docker

En este proyecto utilizaremos una herramienta llamada Docker

En caso de nunca haberla escuchado, a continuación hay un breve resumen y en la siguiente página los pasos para instalar todo lo necesario

### ¿Qué problema resuelve Docker?
A la hora de instalar un programa en nuestro computador, muchas veces ocurre algo como esto:
- El programa necesita una versión específica de Java.

- También necesita ciertas librerías.

- Requiere configuraciones especiales.

- Puede entrar en conflicto con otros programas que ya tienes instalados.

Resultado:
Horas intentando configurar cosas.

Aqui entra Docker permitiendo ejecutar aplicaciones dentro de contenedores
Un contenedor es como una caja cerrada que contiene:
- El programa
- Sus dependencias
- Su configuración
- Todo lo necesario para que funcione
